```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x1, y1 = [0,0], [1.055,1.115]
x2, y2 = [1,1], [1.091,1.180]
x3, y3 = [2,2], [1.025,1.125]

plt.plot(x1, y1, marker = '_', color='black')
plt.plot(0,1.084,marker='^', color='black')
plt.annotate('Lag0-7', xy=(0, 1.17),  xycoords='data',xytext=(0.135, 0.48), textcoords='axes fraction')

plt.plot(x2, y2, marker = '_', color='black')
plt.plot(1,1.135,marker='^', color='black')
plt.annotate('Lag0-14', xy=(0, 1.19),  xycoords='data',xytext=(0.465, 0.58), textcoords='axes fraction')

plt.plot(x3, y3, marker = '_', color='black')
plt.plot(2,1.074,marker='^', color='black')
plt.annotate('Lag0-21', xy=(0, 1.19),  xycoords='data',xytext=(0.8, 0.5), textcoords='axes fraction')

plt.axhline(y=1,linestyle='dashed',color='black')

plt.ylim(0.8,1.5)
plt.ylabel('Relative Risk (RR)',
          fontdict=dict(size=13))
plt.xticks((-0.5,0, 0.5, 1.0, 1.5, 2.0, 2.5), ('', '', '', '','','',''))
fig = plt.gcf()
fig.set_size_inches(10, 8)
fig.savefig('../plots/figure3.png',dpi=200,bbox_inches='tight')
```
